```python
import re
salary = file_data['薪资']
def convert_to_standard_format(text):
    pattern = r'([\d.万千]+)-([\d.万千]+)'
    match = re.search(pattern, text)
    if match:
        lower = match.group(1)
        upper = match.group(2)
        lower = convert_to_wan_format(lower)
        upper = convert_to_wan_format(upper)
        return f"{lower}-{upper}"
    else:
        return text

def convert_to_wan_format(salary):
    if '千' in salary:
        salary = float(salary.replace('千', '')) / 10
    elif '万' in salary:
        salary = float(salary.replace('万', ''))
    else:
        # Handle cases like "130-230/天" or "70-100元/时"
        return salary  # Return original text if it doesn't match expected formats
    return f"{salary:.1f}万"
```

```python
# 处理每条数据
formated_salary = []
for line in salary:
    formatted_line = convert_to_standard_format(line)
    print(formatted_line)
    formated_salary.append(formatted_line)
```

```python
import numpy as np
def preprocess_data(ser):
    # 这里将面议、130-230、70-100等数据替换为NaN
    processed_list = []
    for item in ser:
        if item == "面议" or re.match(r'^\d+-\d+$', item):  # 匹配数字-数字格式
            processed_list.append(np.nan)
        else:
            processed_list.append(item)
    return processed_list

# 预处理数据
processed_data = preprocess_data(formated_salary)
print(processed_data)
```

```python
# 假设我们有一个包含招聘人数数据的Series
recruitment_data = file_data['招聘人数']
# 创建一个新Series来存储转换后的数据
converted_data = pd.Series(np.zeros(len(recruitment_data), dtype=int))

# 遍历原始数据并进行转换
for i, value in recruitment_data.iteritems():
    if "招" in value:
        # 将"招数字人"格式的字符串转换为整数
        number = int(value.replace("招", "").replace("人", ""))
        converted_data[i] = number
    elif "天/周" in value:
        # 将"数字天/周"格式的字符串转换为整数
        number = int(value.replace("天/周", ""))
        converted_data[i] = number

# 打印转换后的Series
converted_data
```

```python
file_data['招聘人数'] = converted_data
```

```python
import pandas as pd

# 读取CSV文件
df = df_filtered

# 提取薪资列
salary_ranges = df['薪资']

# 定义一个函数来计算范围的中点
def calculate_midpoint(salary_range):
    if '万' in salary_range:
        # 去掉万字，并分割范围
        salary_range = salary_range.replace('万', '')
        low, high = map(float, salary_range.split('-'))
        # 返回中点
        return (low + high) / 2
    return None

# 应用函数到每个薪资范围
df['中点'] = salary_ranges.apply(calculate_midpoint)

# 计算中位数
median_salary = df['中点'].median()

print(f"薪资范围的中位数是: {median_salary} 万")

# 将更新后的DataFrame保存回CSV文件
df.to_csv('./data/merged_file.csv', index=False)

print("已将中点添加到CSV文件中。")
```
