```python
import pandas as pd
data = pd.read_csv('./data/merged_file.csv')
data
```

```python
import re
salary = data['薪资']
def convert_to_standard_format(text):
    pattern = r'([\d.万千]+)-([\d.万千]+)'
    match = re.search(pattern, text)
    if match:
        lower = match.group(1)
        upper = match.group(2)
        lower = convert_to_wan_format(lower)
        upper = convert_to_wan_format(upper)
        return f"{lower}-{upper}"
    else:
        return text

def convert_to_wan_format(salary):
    if '千' in salary:
        salary = float(salary.replace('千', '')) / 10
    elif '万' in salary:
        salary = float(salary.replace('万', ''))
    else:
        # Handle cases like "130-230/天" or "70-100元/时"
        return salary  # Return original text if it doesn't match expected formats
    return f"{salary:.1f}万"

# 处理每条数据
formated_salary = []
for line in salary:
    formatted_line = convert_to_standard_format(line)
    print(formatted_line)
    formated_salary.append(formatted_line)
    
```

```python
print(formated_salary)
```

```python
import numpy as np
def preprocess_data(ser):
    # 这里将面议、130-230、70-100等数据替换为NaN
    processed_list = []
    for item in ser:
        if item == "面议" or re.match(r'^\d+-\d+$', item):  # 匹配数字-数字格式
            processed_list.append(np.nan)
        else:
            processed_list.append(item)
    return processed_list

# 预处理数据
processed_data = preprocess_data(formated_salary)
print(processed_data)
```

```python
data['薪资'] = pd.Series(processed_data)
data
```

```python
# 检测重复值记录
data[data.duplicated() == True]
```

```python
# 删除重复记录
data.drop_duplicates(inplace=True)
```

```python
# 删除空值记录
data.dropna(inplace=True)
data
```

```python
data['min_salary'] = data['薪资'].str.split('-').str[0]
data['max_salary'] = data['薪资'].str.split('-').str[1]
data
```

```python
data['min_salary'] = data['min_salary'].apply(lambda x: x[:-1]).astype('float64')
data['max_salary'] = data['max_salary'].apply(lambda x: x[:-1]).astype('float64')
data.head()
```

```python
data.rename(columns={'min_salary': 'min_salary(万元)', 'max_salary': 'max_salary(万元)'}, inplace=True)
data.tail()
```

```python
data.reset_index(drop=True, inplace=True)
data.tail()
```

```python
data['avg_salary(万元)'] = (data['min_salary(万元)'] + data['max_salary(万元)'])/2
data.head()
```

```python
# 保存任务一结果为job_data_avg.csv
data.to_csv('./data/job_data_avg.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# 经验要求
experience_levels = ['经验不限', '3-5年', '1-3年', '5-10年', '无经验']

# 岗位数量和招聘人数数据
job_counts = [109, 93, 83, 23, 8]
recruitment_counts = [1687, 267, 453, 46, 266]

# 创建柱状图
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('经验要求')
ax1.set_ylabel('岗位数量', color=color)
ax1.bar(experience_levels, job_counts, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('招聘人数', color=color)
ax2.bar(experience_levels, recruitment_counts, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('不同经验要求下的岗位数量和招聘人数')

plt.show()
```

```python
# 1、用人工智能相关岗位关键技能生成词云图：
from wordcloud import WordCloud
from collections import Counter

result = ''.join(data['岗位关键技能'].astype(str))
words = result.split('-')
# 过滤掉空字符串
words = [word for word in words if word]
# 统计每个词出现的频率
word_counts = Counter(words)
# print(word_counts)

# 创建词云对象
wordcloud = WordCloud(font_path='./simhei.ttf', width=800, height=400, 
                      background_color='white', min_font_size=10).generate_from_frequencies(word_counts)

# 显示词云
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
